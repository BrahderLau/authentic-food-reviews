```python
# Import all the necessary python libraries
import pandas as pd
import numpy as np
```

```python
# Initialize Variables
file_path = 'premis_makanan_list_part_3_final.csv'
df = pd.read_csv(file_path, encoding='utf-8')
display (df.head(10))
df.shape
```

```python
# Remove white space beginning and at the end
# Make all characters CAPITALIZE using upper function
for column in df.select_dtypes(include=['object']).columns:
    # Strip whitespace and capitalize all characters
    df[column] = df[column].str.strip().str.upper()
```

```python
# Remove prefixes (any number followed by a space)
df['Premise Name'] = df['Premise Name'].str.replace(r'^#\d+\s*', '', regex=True)

# Remove all special characters except ' @ & and -
df['Premise Name'] = df['Premise Name'].str.replace(r"[^a-zA-Z0-9\s'@&-.()]", "", regex=True)
```

```python
for column in df.columns:
    max_value = max(df[column].astype(str), key=len)  # Find the value with the maximum length
    max_length = len(max_value)  # Calculate the length of this value
    min_value = min(df[column].astype(str), key=len)  # Find the value with the maximum length
    min_length = len(min_value)  # Calculate the length of this value
    print(f"Column '{column}' has the shortest number of characters of {min_length} characters with the value of '{min_value}'.")
    print(f"Column '{column}' has the longest number of characters of {max_length} characters with the value of '{max_value}'.")
```

```python
df.loc[df['Premise Name'].str.len() == 1, 'Premise Name'] = None
df.loc[df['Address'].str.len() == 1, 'Address'] = None
```

```python
df_cleaned = df.dropna(subset=['Premise Name'])
```

```python
display (df_cleaned.info())
display (df_cleaned.describe())
display (df_cleaned.isnull().sum())
```

```python
df_null_date = df_cleaned[df_cleaned['Halal Certification Expiry Date'].isna()]
display (df_null_date)
```

```python
premise_counts = df_cleaned.groupby('Premise Name').size().reset_index(name='Counts')
display( premise_counts)

num_unique_rows_of_duplicates = (premise_counts['Counts'] > 1).sum()
display( num_unique_rows_of_duplicates)

total_duplicates_count = (premise_counts['Counts'] - 1).sum()
display( total_duplicates_count)

# Step 2: Merge this count back into the original DataFrame
df_with_counts = pd.merge(df_cleaned, premise_counts, on='Premise Name', how='left')

# Step 3: Sort the DataFrame by 'Premise Name'
df_with_counts_sorted = df_with_counts.sort_values(by='Premise Name')

display( df_with_counts_sorted.tail(10))
```

```python
# Group premises with the similar name: With or without 'S

# Convert 'Halal Certification Expiry Date' to datetime, specifying the format and handling invalid dates by converting them to NaT
df_cleaned['Halal Certification Expiry Date'] = pd.to_datetime(df_cleaned['Halal Certification Expiry Date'], format='%Y-%m-%d', errors='coerce')

# Normalize 'Premise Name' for grouping
df_cleaned['Normalized Premise Name'] = df_cleaned['Premise Name'].str.replace("'S", "", regex=False)

# Sort the DataFrame by 'Normalized Premise Name' and then 'Halal Certification Expiry Date' in descending order to ensure the latest dates come first
df_cleaned_sorted = df_cleaned.sort_values(by=['Normalized Premise Name', 'Halal Certification Expiry Date'], ascending=[True, False])

# Define a function to apply the selection logic within each group
def select_record(group):
    if group['Premise Name'].str.contains("'S").any():
        # Prefer records with 'S, and since the group is already sorted by date, take the first
        preferred = group[group['Premise Name'].str.contains("'S")].head(0)
        if not preferred.empty:
            return preferred
    # If no records with 'S or we're in a group without 'S, return the first row since it's the latest date
    return group.head(1)

# Group by 'Normalized Premise Name' and apply the selection function
df_grouped = df_cleaned_sorted.groupby('Normalized Premise Name', as_index=False).apply(select_record)

# Drop the 'Normalized Premise Name' column and reset the index
df_final = df_grouped.drop(columns=['Normalized Premise Name']).reset_index(drop=True)

# Display the size and df
display(df_final.shape)
display(df_final.head()) 
```

```python
display (df_final.info())
display (df_final.describe())
display (df_final.isnull().sum())
display (df_final['Premise Name'].nunique())
```

```python
df_null_address = df_final[df_final['Address'].isnull()]
display(df_null_address)
```

```python
# Dropping null address because the halal certificate is expired anyway
df_final = df_final.dropna(subset=['Address'])
```

```python
df_final.to_csv('premis_makanan_list_part_6_cleaned.csv', index=False)
```
