```python
# Import all the necessary python libraries
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Define CONSTANTS
BASE_URL = 'https://www.halal.gov.my/v4/index.php?data=ZGlyZWN0b3J5L2luZGV4X2RpcmVjdG9yeTs7Ozs='
CATEGORY_SYARIKAT = '&category=Syarikat'
CATEGORY_PREMIS_MAKANAN = '&category=PE'
PAGE_QUERY_PARAMETER = '&page='
CARI_QUERY_PARAMTER = '&cari='
```

```python
# Initialize Variables
file_path = 'fnb_list_part_2_cleaned.csv'
df = pd.read_csv(file_path, encoding='utf-8')
df.head(10)
```

```python
test_full_name = df['Name'].iloc[227].replace(' ', '%20')
test_simplified_name = df['Simplified Name'].iloc[227].replace(' ', '%20')
test_simplified_name
```

```python
url = (
       BASE_URL 
       + CATEGORY_PREMIS_MAKANAN
       + CARI_QUERY_PARAMTER + test_simplified_name
)

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

print(soup.prettify())
```

```python
# Find the <b> tag that contains the total record count
total_record_tag = soup.find('b', string=lambda text: 'Total Record :' in text)

total_records = 0
number_of_pages = 1
if total_record_tag:
    try:
        # Extract the number after 'Total Record :'
        total_records = int(total_record_tag.text.split('Total Record :')[1].split('-')[0].strip())
        number_of_pages = int(total_record_tag.text.split('From')[1].strip())
        
    except ValueError:
        # In case the number extraction fails
        print("Failed to extract the total record count.")

# Check if the total record is 0
if total_records == 0:
    print("Total record is 0.")
else:
    print(f"Total record is {total_records}.")
    print(f"Number of pages is {number_of_pages}.")
```

```python
def do_menu1(url, cari_value):
    # Replace "&" with "@d" in the cari value
    gcari = cari_value.replace("&", "@d")
    
    # Construct the full URL including the 'cari' parameter
    full_url = f"{url}&cari={gcari}"
    
    # Send a request (assuming it's a POST request; adjust if it's GET)
    # Note: If the method should be GET, use params=payload in requests.get()
    response = requests.post(full_url)
    
    # Assuming you want to do something with the response
    print(response.text)

do_menu1('https://www.halal.gov.my/v4/index.php?data=ZGlyZWN0b3J5L2luZGV4X2RpcmVjdG9yeTs7Ozs=&amp;negeri=&amp;category=PE','bananabro')
```
