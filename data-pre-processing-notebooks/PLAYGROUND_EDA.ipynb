```python
import pandas as pd
import numpy as np
import ast
```

```python
file_path = '../csv_data/fnb_list_part_7_reviews.csv'
df = pd.read_csv(file_path, encoding='utf-8')
display(df['Category'].isna().sum())
display(df)
```

```python
import ast

# Assuming df is your DataFrame and 'Category' is the column containing the string representation of a list
category_string = df['Category'].iloc[0]

# Parse the string representation of the list into an actual list
category_list = ast.literal_eval(category_string)

# Print the resulting list
print(category_list)
```
