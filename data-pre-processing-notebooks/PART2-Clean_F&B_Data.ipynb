```python
# Initialize Variables
file_path = 'csv_data/fnb_list_part_2_corrected.csv'
df = pd.read_csv(file_path, encoding='utf-8')
display (df.shape)
display (df.head())
```

```python
# Since FNB is unique in Google Map, there is no need to convert everything to upper and merge the duplicates

agg_funcs = {'Category': lambda x: list(set(x))} 
other_columns = [col for col in df.columns if col not in ['Name', 'Category']]

# For other columns, we choose the first occurrence
for col in other_columns:
    agg_funcs[col] = 'first'

df_grouped = df.groupby('Name', as_index=False).agg(agg_funcs)
display (df_grouped)
```

```python
def remove_halal_food(categories):
    """Remove 'Halal Food' from the categories list. Return None if empty after removal."""
    filtered_categories = [category for category in categories if category != 'Halal Food']
    if not filtered_categories:  # Check if the list is empty
        return None  # Return None if there are no categories left
    return filtered_categories

# Apply the updated function to each row's category list in the 'Category' column
df_grouped['Category'] = df_grouped['Category'].apply(remove_halal_food)
```

```python
df = df_grouped
# Display the first few rows to verify the change
display( df.shape)
display( df.head())
```

```python
# Detect duplicates

counts = df['Name'].value_counts()
duplicates = counts[counts > 1]
print(duplicates.sum())
duplicates = counts[counts > 1].reset_index().values.tolist()
duplicates_df = pd.DataFrame(duplicates, columns=['Name', 'Count'])
duplicates_df.to_csv('csv_data/fnb_list_part_2_duplicates.csv', index=False)
if duplicates:
    print(len(duplicates))
    for name, count in duplicates:
        print(f'"{name}", {count}')
```

```python
for column in df.columns:
    min_value = min(df[column].astype(str), key=len)
    min_length = len(min_value)
    max_value = max(df[column].astype(str), key=len)
    max_length = len(max_value)
    print(f"Column '{column}' has the shortest number of characters of {min_length} characters with the value of '{min_value}'.")
    print(f"Column '{column}' has the longest number of characters of {max_length} characters with the value of '{max_value}'.")
```

```python
# Testing

filtered_df = df[df['Name'].str.contains(r"[^\w\s\-.&'!]|_", regex=True)]
print(len(filtered_df))
print(filtered_df['Name'])
```

```python
# Testing

cleaned_df = filtered_df.copy()
cleaned_df['Name'] = cleaned_df['Name'].str.replace(r"[^\w\s\-.&'!]+.*", '', regex=True)
print(cleaned_df['Name'])
```

```python
# Add the 'Simplified Name' column by cleaning the 'Name' column
df['Simplified Name'] = df['Name'].str.replace(r"[^\w\s\-.&'!]+.*", '', regex=True)

# Get a list of all column names after adding the new column
cols = list(df.columns)

# Find the index of 'Name' to calculate the position right after 'Name'
name_index = cols.index('Name') + 1

# Reorder columns by ensuring 'Simplified Name' is right after 'Name' without duplication
# Make sure to exclude the original 'Simplified Name' from its previous position
cols = [col for col in cols if col != 'Simplified Name']  # Remove 'Simplified Name' if it exists elsewhere
cols.insert(name_index, 'Simplified Name')  # Insert 'Simplified Name' right after 'Name'

# Reassign reordered columns to the DataFrame
df = df[cols]
df.head(10)
```

```python
df.to_csv('csv_data/fnb_list_part_2_cleaned.csv',index=False,encoding='utf-8')
```
