```python
# Import all the necessary python libraries
import re
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
```

```python
# Define CONSTANTS
BASE_URL_WITH_CARI_PARAMETER = 'https://www.halal.gov.my/v4/index.php?data=ZGlyZWN0b3J5L2luZGV4X2RpcmVjdG9yeTs7Ozs=&cari='
SYARIKAT_BTN_XPATH = '/html/body/form/div/div/div[1]/div/ul/li[2]/a'
PREMIS_MAKANAN_BTN_XPATH = '/html/body/form/div/div/div[1]/div/ul/li[4]/a'
PREMIS_MAKANAN_LIST_XPATH = '//li[@class="clearfix search-result-data"]'
TOTAL_RECORD_XPATH = '//b[contains(text(), "Total Record")]'
```

```python
# Initialize Variables
file_path = 'csv_datanb_list_part_2_cleaned.csv'
df = pd.read_csv(file_path, encoding='utf-8')
df.head(10)
```

```python
test_full_name = df['Name'].iloc[226].replace(' ', '%20')
test_simplified_name = df['Simplified Name'].iloc[226].replace(' ', '%20')
test_simplified_name
```

```python
driver = webdriver.Safari()
driver.maximize_window()
url = BASE_URL_WITH_CARI_PARAMETER + test_simplified_name
driver.get(url)
```

```python
# SEARCH PREMIS MAKANAN FIRST
# USE THE FULL NAME, THEN RECONFIRM WITH ADDRESS
# USE THE SIMPLIFIED NAME, THEN RECONFIRM WITH ADDRESS
# USE THE SYARIKAT NAME FROM THE RESULT
# HALAL CERT - YES
# HALAL STATUS - DITARIK BALIK
# HALAL SIJIL - DATE
# HALAL SYARIKAT NAME
# HALAL SYARIKAT ADDRESS
```

```python
premis_makanan_btn_element = WebDriverWait(driver, 30).until(
    EC.presence_of_element_located((By.XPATH, PREMIS_MAKANAN_BTN_XPATH))
)

bil_premis_makanan = premis_makanan_btn_element.text

# Check if the premis makanan btn contains any digit
if re.search(r'\d', bil_premis_makanan):
    premis_makanan_btn_element.click()
```

```python
total_record_element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, TOTAL_RECORD_XPATH))
)

print(total_record_element.text)

premis_makanan_list_elements = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.XPATH, PREMIS_MAKANAN_LIST_XPATH))
)

if len(premis_makanan_list_elements) == 1:
    driver.find_element(By.XPATH, PREMIS_MAKANAN_LIST_XPATH)

for element in premis_makanan_list_elements:
    print('test') # Get the text of the current element
len(premis_makanan_list_elements)
```
